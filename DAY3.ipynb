```python
# Input: number of white and black balls
white = int(input("Enter the number of white balls: "))
black = int(input("Enter the number of black balls: "))

row = 1
color = "white"  # Start with white for the first row
height = 0       # To count how many rows we can build

while True:
    if color == "white":
        if white >= row:
            white -= row
            height += 1
            color = "black"  # Alternate the color
            row += 1
        else:
            break
    else:  # color is black
        if black >= row:
            black -= row
            height += 1
            color = "white"
            row += 1
        else:
            break

# Output the result
print("The maximum height of the triangle is:", height)
```
