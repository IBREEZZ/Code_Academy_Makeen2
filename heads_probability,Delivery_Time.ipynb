# **heads probability**

```python
import random
import matplotlib.pyplot as plt

heads = 0 #how many times heads comes
prob = [] #stores the probability of heads after each toss

for i in range(1, 1501): #1500 coin tosses
    if random.choice(['H', 'T']) == 'H': #randomly  either 'H' (heads) or 'T' (tails)
        heads += 1   #If  heads increment the heads counter +1
    prob.append(heads / i) #After each toss, calculate heads / total tosses so far #add to prob

#line chart
plt.plot(prob)
plt.axhline(0.5, color='red', linestyle='--')
plt.title('Probability of Heads Over Time')
plt.xlabel('Number of Tosses')
plt.ylabel('P(Heads)')
plt.show()
```


# **Delivery Time**


```python
import random

# Number of orders
num_orders = 5

# List to store total time of each order
total_times = []

for i in range(num_orders):
    decision = random.randint(3, 20)
    confirmation = random.randint(1, 15)
    preparation = random.randint(10, 45)
    delivery = random.randint(5, 30)

    total = decision + confirmation + preparation + delivery
    total_times.append(total)

    print(f"Order {i+1}: Total Time = {total} minutes")

# Calculate and print the average
average = sum(total_times) / num_orders
print(f"\nAverage Delivery Time: {average:.2f} minutes")
```
