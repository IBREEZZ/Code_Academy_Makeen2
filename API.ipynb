```python
import requests

# Step 1: Define the API URL
url = "https://fakestoreapi.com/products"

# Step 2: Make a GET request to fetch the data
try:
    response = requests.get(url)
    response.raise_for_status()  # Raise error for bad responses
    products = response.json()

    # Step 3: Display each product
    for product in products:
        print(f"Title: {product['title']}")
        print(f"Price: ${product['price']}")
        print('-' * 40)

except requests.exceptions.RequestException as e:
    print("Error fetching data:", e)
```

```python
from IPython.display import display, HTML

html_code = """<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <title>About Me & Products</title>
  <style>
    body {
      font-family: Arial, sans-serif;
      padding: 20px;
      background: #f9f9f9;
    }
    nav a {
      margin-right: 15px;
      cursor: pointer;
      color: blue;
      text-decoration: underline;
    }
    .hidden {
      display: none;
    }
    .product {
      background: white;
      border: 1px solid #ddd;
      border-radius: 6px;
      padding: 15px;
      margin-bottom: 15px;
      box-shadow: 0 1px 4px rgba(0,0,0,0.1);
    }
    .product-title {
      font-weight: bold;
      font-size: 1.1em;
      margin-bottom: 6px;
    }
    .product-price {
      color: #2c7;
      font-weight: bold;
    }
    #error {
      color: red;
      font-weight: bold;
    }
  </style>
</head>
<body>

  <nav>
    <a id="aboutLink">About Me</a>
    <a id="productsLink">Products</a>
  </nav>

  <div id="aboutSection">
    <h1>About Me</h1>
    <p>Hello! I am a product enthusiast who loves sharing cool products with you.</p>
  </div>

  <div id="productsSection" class="hidden">
    <h1>Fake Store Products</h1>
    <div id="products-container">Loading products...</div>
    <div id="error"></div>
  </div>

  <script>
    const aboutLink = document.getElementById('aboutLink');
    const productsLink = document.getElementById('productsLink');
    const aboutSection = document.getElementById('aboutSection');
    const productsSection = document.getElementById('productsSection');
    const container = document.getElementById('products-container');
    const errorDiv = document.getElementById('error');

    aboutLink.addEventListener('click', () => {
      aboutSection.classList.remove('hidden');
      productsSection.classList.add('hidden');
    });

    productsLink.addEventListener('click', () => {
      aboutSection.classList.add('hidden');
      productsSection.classList.remove('hidden');
      // Fetch products if not loaded yet
      if (!productsSection.dataset.loaded) {
        fetchProducts();
      }
    });

    function fetchProducts() {
      fetch('https://fakestoreapi.com/products')
        .then(response => {
          if (!response.ok) throw new Error(`HTTP error! status: ${response.status}`);
          return response.json();
        })
        .then(products => {
          container.innerHTML = '';
          products.forEach(product => {
            const productDiv = document.createElement('div');
            productDiv.className = 'product';

            const title = document.createElement('div');
            title.className = 'product-title';
            title.textContent = product.title;

            const price = document.createElement('div');
            price.className = 'product-price';
            price.textContent = `$${product.price}`;

            productDiv.appendChild(title);
            productDiv.appendChild(price);
            container.appendChild(productDiv);
          });
          productsSection.dataset.loaded = 'true'; // Mark as loaded to avoid re-fetching
        })
        .catch(err => {
          container.innerHTML = '';
          errorDiv.textContent = 'Error fetching data: ' + err.message;
        });
    }

    // Optional: show About Me by default on page load
    aboutSection.classList.remove('hidden');
    productsSection.classList.add('hidden');
  </script>

</body>
</html>
"""

display(HTML(html_code))
```
