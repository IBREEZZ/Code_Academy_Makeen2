```python
# Get the height of the triangle from the user
height = int(input("Enter the height of triangle: "))

# Loop through each row from 1 to height
for i in range(1, height + 1):
    # Print spaces to center the triangle
    print("  " * (height - i), end="")

    # Start with 1 for the increasing numbers
    num = 1
    # Print the increasing numbers (1, 2, 4, 8, ...)
    for j in range(i):
        print(num, end=" ")
        num *= 2  # Double the number

    # Step back one level to start decreasing
    num //= 2
    # Print the decreasing numbers (..., 4, 2, 1)
    for j in range(i - 1):
        num //= 2  # Divide by 2
        print(num, end=" ")

    # Move to the next line
    print()
```
